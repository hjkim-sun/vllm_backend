# 1. 모델 로드

```python
from transformers import AutoConfig, AutoModelForCausalLM

# config.json 파일만 로드하고 실제 가중치는 로드하지 않음
config = AutoConfig.from_pretrained("openai-community/gpt2-xl", trust_remote_code=True)
model = AutoModelForCausalLM.from_config(
    config,
    trust_remote_code=True
).to('cuda')

```

# 2. 모델 구조 보기

```python
print(model)
```

```python
print(model.transformer.h[4].attn.c_attn)
print(model.transformer.h[4].attn.c_attn.weight)
```
