# 2. 모델 로드
 - 모델 로드 중 실패 발생시 제보 부탁드립니다. (Colab 기본 라이브러리 버전과 충돌 발생 가능성 있음)

```python
from vllm import LLM, SamplingParams
```

```python
# 허깅페이스에 있는 모델을 찾아 다운로드
llm = LLM(
    model='openai-community/gpt2-xl'
)
```

# 3. 추론 수행

```python
sampling_params = SamplingParams(
    temperature=0.1,
    top_p=0.95,
    max_tokens=200,
    #do_sample=True
)
```

```python
outputs = llm.generate(['Hello, who are you?'], sampling_params)
print(outputs)
```

```python
for output in outputs:
  print(f'request_id: {output.request_id}')
  print(f'Prompt: {output.prompt!r}')
  print(f'prompt_token_ids: {output.prompt_token_ids}')
  print(f'Generated text: {output.outputs[0].text!r}')
  print(f'token_ids text: {output.outputs[0].token_ids}')
```
