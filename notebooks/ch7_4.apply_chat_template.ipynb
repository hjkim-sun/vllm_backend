# 1. 토크나이저 로드

```python
from transformers import AutoTokenizer

# 허깅페이스에서 tokenizer.json 파일을 다운로드하여 로드
tokenizer = AutoTokenizer.from_pretrained('openai/gpt-oss-20b')
```

# 2. apply_chat_template 수행

```python
messages = [
    {
      "role": "system", "content": "You are a helpful assistant."
    },
    {
      "role": "user", "content" : "Hello, how are you?"
    }
]
encoded_input = tokenizer.apply_chat_template(
    messages,
    return_tensors='pt',
    add_generation_prompt=True,
    ).cuda('cuda')
print(encoded_input)
```

```python

token_ids = encoded_input[0].tolist()
token_str = tokenizer.convert_ids_to_tokens(encoded_input[0])

print(token_str)
```

- 토큰 ID와 문자열 묶어서 같이 보기 

```python
for tid, tstr in zip(token_ids, token_str):
    print(f"{tid:>8}  {tstr}")
```
